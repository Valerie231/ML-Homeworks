  3.1. без використання жодних бібліотек, користуючись тільки Python.

Виведіть результат обчислень в кожному з випадків на екран. Порівняйте результати зі значенням вибіркового середнього. Напишіть коментар, для яких задач ви б використовували вибіркове середнє, а для яких - медіану?

```python
sorted_list = sorted(sum_purchases)

#checking if the number of elements is odd or even
n = len(sorted_list)
if n % 2 == 1:
    #If odd, median is the middle number
    median = sorted_list[n // 2]
else:
    #If even, median is the average of the two middle numbers
    median = (sorted_list[n // 2 - 1] + sorted_list[n // 2]) / 2

print("Median using python:", median)
print("Median using numpy:", np.median(sum_purchases))
```

**Завдання 4**. Обчислість стандартне відхилення наведеної вибірки в змінній `sum_purchases` кожним з зазначених методів:  

  4.1. без використання жодних бібліотек, користуючись тільки Python.  
  4.2. використовуючи відповідний метод бібліотеки numpy.  

Виведіть результат обчислень в кожному з випадків на екран.

```python
# Calculating the mean of the list
mean = sum(sum_purchases)/len(sum_purchases)

# Subtracting the mean from each element
deviations = [i - mean for i in sum_purchases]

# Squaring each deviation
sqr_deviations = [i ** 2 for i in deviations]

# Calculating the mean of the squared deviations
mean_sqr_deviations = sum(sqr_deviations)/len(sqr_deviations)

# Taking the square root
std_dev = mean_sqr_deviations ** 0.5

# Print the standard deviation
print("Standard deviation using python:", std_dev)

#Calculating standard deviation using numpy:
print("Standard deviation using numpy:", np.std(sum_purchases))
```

```python
visits = [ 6,  2, 10,  3,  9,  9,  2,  3,  6,  6, 10,  2,  5, 10,  3,  5,  4,
        7,  4,  2]
```

```python
# Calculating means of visits and purchases 
mean_vs = sum(visits) / len(visits)
mean_pr = sum(sum_purchases) / len(sum_purchases)

# Calculating deviations
deviations_vs = [i - mean_vs for i in visits]
deviations_pr = [i - mean_pr for i in sum_purchases]

# Calculating the product of deviations
product_deviations = [v * p for v, p in zip(deviations_vs, deviations_pr)]

# Calculating the sum of products of deviations
sum_product_deviations = sum(product_deviations)

# Calculating the sum of squared deviations visits and purchases
sum_squared_deviations_v = sum(v ** 2 for v in deviations_vs)
sum_squared_deviations_p = sum(p ** 2 for p in deviations_pr)

# Taking the square root of the sum of squared deviations
sqrt_sum_squared_deviations_v = (sum_squared_deviations_v) ** 0.5
sqrt_sum_squared_deviations_p = (sum_squared_deviations_p) ** 0.5

# Calculating the Pearson correlation coefficient
pearson_coefficient = sum_product_deviations / (sqrt_sum_squared_deviations_v * sqrt_sum_squared_deviations_p)

print("Pearson correlation coefficient:", pearson_coefficient)
print("Pearson correlation coefficient using numpy:", np.corrcoef(visits, sum_purchases)[0, 1])
```

Давайте вилучимо записи, які відповідають значенням з викидами в сумі покупок і обчисліть коеф. кореляції. Для цього поставте `threshold`, який дорінює середньому значенню + 2 стандартних вдіхилля суми покупок і всі значення, які більше за нього, будемо вважати викидами (є і інші правила для визначення викидів - з ним и познайомимось пізніше).

Вилучіть значення з обох масивів і обчисліть коеф. кореляції Пірсона з допомогою `numpy` для кількості візитів і суми покупок без записів-викидів. Виведіть результат на екран, побудуйте діаграму розсіювання і проаналізуйте результат: що можемо сказати про залежність в даних?

```python
def threshold(x):
    return np.mean(x) + 2*np.std(x)

sum_purchases_without_outliers = np.array(sum_purchases)[sum_purchases < threshold(sum_purchases)]

# Removing the corressponding values from the visits array
visits_visout_outliers = np.array(visits)[sum_purchases < threshold(sum_purchases)]

print(np.corrcoef(sum_purchases_without_outliers, visits_visout_outliers)[0,1])
```

```python
plt.scatter(sum_purchases_without_outliers, visits_visout_outliers)
plt.title('Scatter Plot of Purchases vs. Visits')
plt.xlabel('Sum of Purchases')
plt.ylabel('Number of Visits');
```

There is a slightly negative correlation between the sum of purchases and number of visits of each customer, which could be connected to the fact that people who live nearby come to the store more often and therefore make smaller purchases, and those who visit the store rarely usually buy a lot; however, if we check the purchases data - it doesn't differ much, therefore I would suggest that in this case correlation doesn't mean causation - it can be just a random uninformative result caused by the very small sample :)
